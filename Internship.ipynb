```python
import pandas as pd

# Load movies.csv, ratings.csv, and links.csv into DataFrames
movies_df = pd.read_csv(r"C:\Users\Administrator\Downloads\movie_data\movies.csv")
ratings_df = pd.read_csv(r"C:\Users\Administrator\Downloads\movie_data\ratings.csv")
links_df = pd.read_csv(r"C:\Users\Administrator\Downloads\movie_data\links.csv")

# Merge movies and links to get IMDb IDs
movies_links = pd.merge(movies_df, links_df[['movieId', 'imdbId']], on='movieId', how='inner')

# Count the number of ratings for each movie
ratings_count = ratings_df.groupby('movieId').size().reset_index(name='num_ratings')

# Filter movies with more than 50 user ratings
movies_over_50_ratings = movies_links.merge(ratings_count[ratings_count['num_ratings'] > 50], on='movieId', how='inner')

# Find the movie with the highest IMDb rating among movies with over 50 ratings
highest_rated_movie = movies_over_50_ratings.loc[movies_over_50_ratings['imdbId'].idxmax()]

highest_rated_movie_id = highest_rated_movie['movieId']

print(f"The movieId of the movie with the highest IMDb rating and more than 50 user ratings is: {highest_rated_movie_id}")
```

```python
import pandas as pd

# Load movies.csv, ratings.csv, and links.csv into DataFrames
movies_df = pd.read_csv(r"C:\Users\Administrator\Downloads\movie_data\movies.csv")
ratings_df = pd.read_csv(r"C:\Users\Administrator\Downloads\movie_data\ratings.csv")
links_df = pd.read_csv(r"C:\Users\Administrator\Downloads\movie_data\links.csv")

# Merge movies and links to get IMDb IDs
movies_links = pd.merge(movies_df, links_df[['movieId', 'imdbId']], on='movieId', how='inner')

# Count the number of ratings for each movie
ratings_count = ratings_df.groupby('movieId').size().reset_index(name='num_ratings')

# Merge movies with more than 50 ratings and links to get Sci-Fi movies with IMDb IDs
scifi_movies = movies_links.merge(ratings_count[ratings_count['num_ratings'] > 50], on='movieId', how='inner')
scifi_movies = scifi_movies[scifi_movies['genres'].str.contains('Sci-Fi', case=False)]

# Find the Sci-Fi movie with the highest IMDb rating
highest_rated_scifi_movie = scifi_movies.loc[scifi_movies['imdbId'].idxmax()]

highest_rated_scifi_movie_id = highest_rated_scifi_movie['movieId']

print(f"The movieId of the Sci-Fi movie with the highest IMDb rating and more than 50 user ratings is:{highest_rated_scifi_movie_id}")
```
